```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
main_train_data = pd.read_csv('./data/train.csv')
main_test_data = pd.read_csv('./data/test.csv')

combined = main_train_data.append(main_test_data, ignore_index=True)
```

```python
print (main_train_data.shape)
print (main_test_data.shape)
print (combined.shape)
print (main_train_data.columns.values)
```

## Pclass

```python
combined['Pclass'].isnull().sum(axis=0)
```

```python
combined[['Pclass', 'Survived']].groupby(['Pclass']).mean()
```

```python
combined.Name.head()
```

```python
combined['Initials'] = combined.Name.map(lambda ini: ini.split(',')[1].split('.')[0].strip())
```

```python
combined['Initials'].value_counts()
```

```python
combined.Initials.loc[ (combined.Initials !=  'Master') & (combined.Initials !=  'Mr') & (combined.Initials !=  'Miss') 
             & (combined.Initials !=  'Mrs')] = 'Others'
```

```python
combined[['Initials', 'Survived']].groupby(['Initials']).mean()
```

```python
combined['Initials'].value_counts()
```

```python
df = pd.concat([combined, pd.get_dummies(combined['Initials'])], axis=1).drop(labels=['Name'], axis=1)
df.head()
```

```python
df.Sex.isnull().sum()
```

```python
df[['Sex','Survived']].groupby(['Sex']).mean()
```

```python
df.Sex = df.Sex.map({'male':0,'female':1})
```

```python
df.head()
```

```python
df.Age.isnull().sum()
```

```python
df['family'] = df['Parch']+df['SibSp']+1
df[['family','Survived']].groupby(['family']).mean()
```

```python
df['family'].value_counts()
```

```python
df.family = df.family.map(lambda x:0 if x>4 else x)
df[['family','Survived']].groupby(['family']).mean()
```

```python
df.head()
```

```python
df.Ticket = df.Ticket.map(lambda x:x[0])
df[['Ticket','Survived']].groupby(['Ticket']).mean()
```

```python
df.head()
```

```python
df['Cabin'].isnull().sum()
```

```python
df = df.drop(labels='Cabin', axis=1)
df.head()
```

```python
df['Embarked'].isnull().sum(axis=0)
```

```python
df.Embarked.fillna('S', inplace=True)
```

```python
df = df.drop(labels='Embarked',axis = 1)
df.head()
```

```python
df['Ticket'] = df['Ticket'].replace(['A','W','F','L','5','6','7','8','9'], '4')
df[['Ticket', 'Survived']].groupby(['Ticket'], as_index=False).mean()
```

```python
df = pd.get_dummies(df,columns=['Ticket'])
df.head()
```

```python
df = df.drop(labels=['SibSp','Parch','Age','Fare','Initials'], axis=1)
Y_train = df[:891]['Survived'].values
X_tran = df[0:891].drop(['Survived', 'PassengerId'], axis=1).values
X_test = df[891:].drop(['Survived', 'PassengerId'], axis=1).values
```

```python
X_tran.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()
model.add(Dense(9,activation='relu',input_shape=(15,)))
model.add(Dense(9,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='relu'))

model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_tran, Y_train, batch_size=32, epochs=50, verbose=1)
```

```python
pred = model.predict(X_test)
final = (pred > 0.5).astype(int).reshape(X_test.shape[0])
```

```python
save = pd.DataFrame({'PassengerId':main_test_data['PassengerId'], 'Survived': final})
```

```python
save.to_csv('output.csv', index=False)
```
